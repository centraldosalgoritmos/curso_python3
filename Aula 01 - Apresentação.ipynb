## 1. História da Linguagem Python

A linguagem Python foi desenvolvida no final da década de 80 pelo holandês, __Guido van Rossum__, enquanto trabalhava no __Centrum Wiskunde & Informatica - CWI__. __Guido__ estava alocado num projeto de uma linguagem chamada __ABC__ que não teve sucesso. Foi então que ele teve a ideia de criar um sucessor para a linguagem de programação __ABC__ bem mais otimizada. __Guido__ planejou gastar poucos meses no desenvolvimento, utilizando seu tempo livre durante o Natal de 1989. 

Porém o desenvolvimento acabou levando mais tempo que o previso e foi somente em Janeiro de 1994 que __Guido__ conseguiu lançar a versão `Python 1.0`. A linguagem Python foi projetada com a filosofia de priorizar a __legibilidade do código__ e __produtividade do desenvolvedor__.
<a id="guido-van-rossum"></a>
<img src="https://miro.medium.com/max/1200/0*zExBvLRUmRRS5m5U.jpg" alt="Guido Van Rossum - 1994" width="70%">

Python é uma linguagem de programação de _alto nível, interpretada, imperativa, orientada a objetos, funcional, de tipagem dinâmica e forte_. 
* _Alto nível_ - São linguagens tem um nível de abstração relativamente elevado, ou seja, sua estrutura é muito longe do código de máquina e bem mais próxima da linguagem humana.
* _Interpretada_ - São linguagens de programação em que o código fonte não é executado diretamente pelo Sistema Operacional e sim por um programa de computador chamado Interpretador.
* _Paradigma Imperativo_ - É um paradigma de programação que descreve a computação como ações, enunciados ou comandos que mudam o estado (variáveis) de um programa.
* _Paradigma Orientado a Objetos_ - É um paradigma de programação baseado na composição e interação entre diversas unidades chamadas de _objetos_.
* _Funcional_ - É um paradigma de programação que trata a computação como uma avaliação de funções matemáticas e que evita estados ou dados mutáveis.
* _Tipagem Dinâmica e Forte_ - Podemos categoriza as linguagens de programação, quanto ao forma de tipagem:
    1. Forte: É quando uma linguagem verifica os tipos de dados utilizados em tempo de execução ou compilação. Ex: Python e Java
    2. Fraca: É quando uma linguagem não verifica os tipos de dados utilizados em tempo de execução ou compilação. Ex: JavaScript e PHP
    3. Estática: É quando uma linguagem exige que  sejam declarados os tipos de dados utilizados, e uma vez definido não são possíveis alterações em tempo de execução. Ex: C# e Java
    4. Dinâmica: É quando uma linguagem não exige que sejam declarados os tipos de dados utilizados e temos a possibilidade de alterar o tipo da nossa variável em tempo de execução. Ex: JavaScript e Python.

Devido às suas características, a linguagem Python é atualmente utilizada em diversos cenários sendo portanto considerada pelo site [Stack Overflow](https://insights.stackoverflow.com/survey/2019#most-loved-dreaded-and-wanted) a __4º Tecnologia mais popular__, a __2ª linguagem "mais amada"__ e a __1ª linguagem "mais requerida"__, além de estar entre as __5 linguagens mais populares__, de acordo com uma pesquisa conduzida pela [RedMonk em Junho de 2019](https://redmonk.com/sogrady/2019/07/18/language-rankings-6-19/).

Um curiosidade é que o nome Python não teve a sua origem relacionada com a espécie de serpentes e sim com o grupo humorístico britânico __Monty Python__, criadores do programa __Monty Python's Flying Circus__.

## 5. Preparando o ambiente

Antes de começar a programar em Python, você precisa ter acesso a uma implementação do __Interpretador__ Python. Um __Interpretador__ é um programa capaz de ler um código fonte e traduzí-lo para uma versão executável específica para o ambiente de execução (seu computador). Há divesas opções para isso:

### 5.1 Utilizar Interpretadores Online
* [Python.org Online Console](https://www.python.org/shell).
* [Repl.it](https://repl.it/languages/python3).
* [Online GBD](https://www.onlinegbd.com/online_python_compiler).
* [Trinket.io](https://trinket.io/python).

### 5.2 Instalar um Interpretador no seu computador

#### 5.2.1 Windows

Normalmente não possui o interpretador para Python 3 instalado, felizmente o processo de instalação é muito simples e envolve apenas baixar o instalador (adequado 32-bit/64-bit) no site da [Python Software Foundation](https://www.python.org), executá-lo.
* Para confirmar a instalação, abra uma __Janela de Comando__ (`Windows + R`, digite `cmd` e tecle `ENTER`).
* Na __Janela de Comando__ digite `python3 --version` e tecle `ENTER`.

#### 5.2.2 Linux

Há uma grande chance do seu sistema já possuir o interpretador Python instalado, mas provavelmente não será a versão mais atual. 
* Para verificar qual versão instalada basta abrir uma __Janela Terminal__ (`Ctrl+T`) e digite `python --version` e tecle `ENTER`.
* Se a versão exibida for diferente de __Python 3.7.4__ você não possui a versão mais atual lançada até a data em que escrevo esta aula e terá que atualizá-la.
* Para atualizar sua versão recomendo:
    * Este tutorial [Instalando Python - RealPython](https://realpython.com/installing-python).
    * Ou este vídeo [Como atualizar a versão do Python no Ubuntu](https://www.youtube.com/watch?v=vizOkM02TMQ)

#### 5.2.3 Mac OS

Uma abordagem bastante recomendada pela comunidade é a instalação utilizando o gerenciador de pacotes __Homebrew__ conforme o tutorial [Instalando Python3 no Mac OS X - Docs Python](https://docs.python-guide.org/starting/install3/osx/).


#### 5.2.4 Instalar a Plataforma Anaconda

Plataforma de Data Science e Machine Learning em Python/R. Disponível para Linux, Windows e Mac OS X. Download e instalação pelo site da plataforma [Anaconda](https://www.anaconda.com/distribuition/).<a id="anaconda"></a>

* Traz diverso pacotes pré-instalados, além do __Jupyter Notebook__ e __Spyder IDE__.
* Gerenciamento e download de módulos/bibliotecas.
* Possibilita o desenvolvimento e treino de modelos de __Machine Learning__ utilizando `scikit-learn`, `TensorFlow` e `theano`.
* Facilita a __Análise de dados, escalabilidade e performance__ utilizando `dask`, `numPy`, `pandas` e `numba`.
* Permite a __Visualização__ de resultados com `matplotlib`, `bokeh`, `datashader` e `holoviews`.

## 7. The Zen of Python

__The Zen of Python__ é uma lista de 19 (talvez exista um 20º) princípios gerais para desenvolvimento.

```python
import this
```

Vamos entender tais princípios:
1. _"Bonito é melhor que feio"_ - Em Python, quando existem várias formas para chegar numa solução, escolhemos a __visualmente mais agradável__.
1. _"Explicito é melhor que implicito"_ - Em Python, sempre preferimos a opção mais __legível__, ou seja, aquela que apresenta todos os elementos para se fazer entender.
1. _"Simples é melhor que complexo"_ - Ao desenvolver um código Python, não busque o caminho mais complexo! Tente manter __simples__!
1. _"Complexo é melhor que complicado"_ - Quando a alternativa simples não é possível, fuja da solução complicada.
1. _"Linear é melhor do que aninhado"_ - Evite criar estruturas dentro de estruturas ... Opte pela alternativa __linear__, sem aninhá-las.
1. _"Esparso é melhor que denso"_ - Não tente amontoar código!
1. _"Legibilidade conta"_ - Ao terminar de desenvolver, verifique a __legibilidade__ do seu código.
1. _"Casos especiais não são especiais o bastante para quebrar as regras. Ainda que praticidade vença a pureza"_ - __Nenhum caso__ é tão especial a ponto de passar por cima das regras!
1. _"Erros nunca devem passar silenciosamente. A menos que sejam explicitamente silenciados"_ - Silenciar exceções é um erro grave de manuntenção de código, pois pode esconder um erro crítico!
1. _"Diante da ambiguidade, recuse a tentação de adivinhar"_ - Mantenha seu código específico, limpo, visualmente agradável e sem ambiguidade.
1. _"Deve haver uma - e preferencialmente apenas uma - maneira óbvia de fazer algo. Embora esse caminho possa não ser óbvio a princípio, a menos que você seja 'Holandês'"_ - Python defende que __deve__ existir apenas uma maneira de se chegar a uma solução. É isso que torna Python __tão simples__. A segunda parte é apenas uma referência ao criador do Python, Guido van Rossum - que é holandês.
1. _"Agora é melhor que nunca. Embora nunca frequentemente seja melhor que já"_ - Primeiramente, não gaste seu tempo demasiadamente com planejamento, as vezes é melhor fazer uma primeira versão e iterar adicionando pequenas melhorias até ter uma versão final. Isso não signigica, entretanto, que que você deve sair fazendo _qualquer coisa_ , apenas pelo fato de se ter algo pronto.
1. _"Se a implementação é difícil de explicar, é uma má ideia"_ - Se a sua solução é de difícil entendimento, revise-a!
1. _"Se a implementação é fácil de explicar, pode ser uma boa ideia"_ - Agora, se a solução é simples de ser explicada, ela __pode ser__ uma boa ideia.
1. _"Namespaces são uma grande ideia - vamos ter mais dessas"_
